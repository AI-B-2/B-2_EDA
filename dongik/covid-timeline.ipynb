```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('./full_grouped.csv')
data
```

```python
data_europe = data[data['WHO Region']=='Europe']
data_europe
```

```python
data_europe.groupby('Date').sum()['New cases'].plot()
plt.title('Europe Timeline')
```

```python
data_africa = data[data['WHO Region']=='Africa']
data_africa.groupby('Date').sum()['New cases'].plot()
plt.title('Africa Timeline')
```

```python
data_em = data[data['WHO Region']=='Eastern Mediterranean']
data_em.groupby('Date').sum()['New cases'].plot()
plt.title('Eastern Mediterranean Timeline')
plt.show()
```

```python
data_am = data[data['WHO Region']=='Americas']
data_am.groupby('Date').sum()['New cases'].plot()
plt.title('Americas Timeline')
plt.show()
```

```python
data_wp = data[data['WHO Region']=='Western Pacific']
data_wp.groupby('Date').sum()['New cases'].plot()
plt.title('Western Pacific Timeline')
plt.show()
```

```python
data_sea = data[data['WHO Region']=='South-East Asia']
data_sea.groupby('Date').sum()['New cases'].plot()
plt.title('South-East Asia Timeline')
plt.show()
```
